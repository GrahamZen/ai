```python
import pandas as pd
```

```python
def convert(fname,col,row):
    st=pd.read_csv(fname)
    d={}
    for i in st.iterrows():
        d[(int(i[1][0]),int(i[1][1]))]=i[1][2]
    f=open('convert'+fname,'w+')
    print(end=',',file=f)
    for j in range(row):
        print(j+1,end=',',file=f)
    print('',file=f)
    for i in range(col):
        for j in range(row):
            p=2**i
            print("{},{}".format(p,d[(p,j+1)]),end='',file=f)
        print('',file=f)
    f.close()
```

```python
d=pd.read_csv('ResNet_DA.csv')
```

```python
d
```

```python
di={}
sl=['static','dynamic','guided']
for st in sl:
    di[st]={}
    for i in range(8):
        p=2**i
        di[st][p]=d[((d['strategy']==st)&(d['thread']==p))].mean()[1]
```

```python
pd.DataFrame(di).to_csv('2.csv')
```
